### 1 ) 데이터 xml 파일 요청

```python
import requests

url = 'http://apis.data.go.kr/1400000/forestStusService/getfirestatsservice'
params ={'serviceKey' : 'Wfv14QgD4D8t2cJ6/xya6tkKk0miuOCHsez2JIyyCFgIe/lpCt5cnk9dN1fjlPcO6N5MYHTJ0lgUiD4X4614xA==', 'numOfRows' : '10000', 'pageNo' : '1', 'searchStDt' : '20160101', 'searchEdDt' : '20210825' }

response = requests.get(url, params=params)
```

### 2 ) 데이터 변환

- xml 불러오기

```python
from bs4 import BeautifulSoup
```

```python
fire_data = BeautifulSoup(response.text, 'html.parser')
```

- 딕셔너리형으로 변환

```python
dict_ = {}

for item in fire_data.find_all('items'):
    for j in item.find_all('item'):
        for k in j:
            dict_[k.name] = list()

for item in fire_data.find_all('items'):
    for j in item.find_all('item'):
        for k in j:
            dict_[k.name].append(k.get_text())
```

```python
dict_.keys()
```

- 판다스 객체로 변환

```python
import pandas as pd
```

```python
def dict_to_pd(dict_):
    df = pd.DataFrame()
    for name, val in dict_.items():
        df = pd.concat([df, pd.DataFrame(val, columns=[name])], axis=1)
    
    return df
```

```python
fire_df = dict_to_pd(dict_)
```

#### 2 ) folium을 이용한 지도 시각화

- 지역 별 지도 시각화를 위해서는 지역별로 경도 위도가 필요하다. 이를 위해 크롤링을 사용한다.

```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://kor.timegenie.com/latitude_longitude/country/kr')
response = response.text
data = BeautifulSoup(response, 'html.parser')
```

```python
lat = []
for i in data.find_all('div'):
    for j in i.find_all('span'):
        lat.append(j.text)
```

```python
lan_lat = {}

for i in data.find_all('div'):
    for idx, j in enumerate(i.find_all('h5')):
        if idx % 3 == 1:
            if j.text == '홈':
                break
            lan_lat[j.text[:2]] = []
            name = j.text[:2]
            
        if (idx % 3 == 0) and idx != 0:
            lan_lat[name] = [float(j.text), float(lat[idx+1])]
        
            
```

```python
lan_lat
```

- 크롤링을 통해 얻은 각 시군구별 좌표를 데이터프레임에 merge시킨다. ( Left Join ) 

```python
df_copy
```

```python
df_lat = pd.DataFrame(lan_lat)
```

```python
df_lat = df_lat.T.reset_index()
```

```python
df_lat.columns = ['locgungu', 'lan', 'lat']
```

```python
df_lat
```

```python
df_copy_1 = pd.merge(left=df_copy, right=df_lat, how='left')
```

```python
df_copy_1
```

```python
df_copy_1
```

```python
df_copy_1.isnull().sum()
```

- 지도 시각화를 위해선 단순히 월별 기준으로 groupby를 진행하는 것이 아니라 시도 또한 groupby에 사용되어야 한다.
